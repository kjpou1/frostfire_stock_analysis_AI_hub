```python
def encode_images_from_directories(directories):
    """
    Encode all images from a list of directories into Base64 strings.

    Parameters:
    - directories (list of str): List of directory paths to scan for images.

    Returns:
    - dict: JSON payload containing Base64-encoded image data.
    """
    # Supported image file extensions
    valid_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".gif"]

    # List to store Base64-encoded images
    base64_images = []

    for directory_path in directories:
        # Ensure the directory exists
        if not os.path.isdir(directory_path):
            print(f"Warning: Directory '{directory_path}' does not exist. Skipping.")
            continue

        # Iterate through all files in the directory
        for filename in os.listdir(directory_path):
            file_path = os.path.join(directory_path, filename)
            # Check if the file has a valid image extension
            if (
                os.path.isfile(file_path)
                and os.path.splitext(filename)[-1].lower() in valid_extensions
            ):
                try:
                    # Read the binary image data and encode it as Base64
                    with open(file_path, "rb") as img_file:
                        base64_string = base64.b64encode(img_file.read()).decode(
                            "utf-8"
                        )
                        base64_images.append(base64_string)
                        #print(f"Encoded: {file_path}")
                except Exception as e:
                    print(f"Failed to encode {file_path}: {e}")

    # Prepare the JSON payload
    return {"base64_images": base64_images}
```

```python
# List of directories to scan
import json


directories = [
    CHARTS_DIR,
    NON_CHARTS_DIR
]

# Generate Base64-encoded images payload
payload = encode_images_from_directories(directories)
print(json.dumps(payload, indent=4))
# # Save the payload to a JSON file (optional)
# with open("encoded_images.json", "w") as json_file:
#     json.dump(payload, json_file, indent=4)

# # Print the payload for inspection
# print(json.dumps(payload, indent=4))
```
